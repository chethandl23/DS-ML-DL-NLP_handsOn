```python
import pandas as pd
import numpy as np
data = pd.read_excel(r"C:\data_science_udemy\DATA\flight_price.xlsx")
data.head()
```

```python
data.info()
```

```python
# Exploratory Data Analysis (EDA) on flight data
# to perform EDA we will take data of journey from the flight_data_eda.ipynb file
data['date'] = data['Date_of_Journey'].str.split('/').str[0].astype(int)
data['month'] = data['Date_of_Journey'].str.split('/').str[1].astype(int)
data['year'] = data['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
#drop the original 'Date_of_Journey' column
data.drop(columns = ['Date_of_Journey'], inplace=True,errors='ignore')
#data.drop('Date_of_Journey', axis=1, inplace=True) can also be used to drop a column
data.head()
```

```python
#removing the unwanted date from 'Arrival_Time' column
data['Arrival_Time'] = data['Arrival_Time'].str.split(' ').str[0]
data.head()
```

```python
data['Arrival_hour'] = data["Arrival_Time"].str.split(":").str[0].astype(int)
data['Arrival_minute'] = data["Arrival_Time"].str.split(":").str[1].astype(int)
data.drop(columns=['Arrival_Time'], inplace=True, errors='ignore')
```

```python
#format departure time
# data['Departure_hour'] = data["Dep_Time"].str.split(":").str[0].astype(int)
# data['Departure_minute'] = data["Dep_Time"].str.split(":").str[1].astype(int)
# data.drop(columns=['Dep_Time'], inplace=True, errors='ignore')
data.head(2)
```

```python
# Fix for extracting hours and minutes from 'Duration' column
def extract_hour(x):
	if 'h' in x:
		return int(x.split('h')[0])
	else:
		return 0

def extract_minute(x):
	if 'm' in x:
		return int(x.split('m')[0].split()[-1])
	else:
		return 0

data['Duration_hour'] = data['Duration'].apply(extract_hour)
data['Duration_minute'] = data['Duration'].apply(extract_minute)
data['duration'] = data['Duration_hour'] * 60 + data['Duration_minute']
data.drop(columns=['Duration', 'Duration_hour', 'Duration_minute'], inplace=True, errors='ignore')
```
